```python
import numpy as np
import pandas as pd
from copy import deepcopy
from bertopic import BERTopic
```

```python
df = pd.read_csv('Train.csv')
```

```python
df.head()
```

```python
df = df[['ABSTRACT']]
```

```python
df.head()
```

Remove Stop Words

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
print(STOP_WORDS)
```

```python
df['ABSTRACT'].apply(lambda x: len([t for t in x.split() if t in STOP_WORDS]))
```

```python
df['ABSTRACT'] = df['ABSTRACT'].apply(lambda x:' '.join([t for t in x.split() if t not in STOP_WORDS]))
```

```python
df.head()
```

Lower case conversion

```python
df['ABSTRACT'] = df['ABSTRACT'].apply(lambda x: x.lower())
```

```python
df
```

Remove Special characters

```python
import re
```

```python
df['ABSTRACT'] = df['ABSTRACT'].apply(lambda x: re.sub('[^A-Z a-z 0-9-]+','',x))
```

```python
df.head()
```

Remove multiple spaces


```python
df['ABSTRACT'] = df['ABSTRACT'].apply(lambda x: ' '.join(x.split()))
```

```python
df.head()
```

```python
docs = list(df.loc[:, "ABSTRACT"].values)
```

```python
docs[:5]
```

```python
model = BERTopic(language="english")
```

```python
topics, probs = model.fit_transform(docs)
```

Extract most freq topics

```python
model.get_topic_freq()
```

```python
model.get_topic(1)
```

Visualize topics

```python
model.visualize_topics()
```
